```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# URL del archivo CSV
url = 'https://raw.githubusercontent.com/Sinchiguano/chicagoUniversity/main/cta_Ridership.csv'

# Leer el archivo CSV en un DataFrame
df = pd.read_csv(url)

# Convertir la columna 'date' al formato datetime
df['date'] = pd.to_datetime(df['date'])

# Establecer la columna 'date' como ndice
df.set_index('date', inplace=True)

# Filtrar por estaciones especficas
Garfield_Dan_Ryan = df[df['stationname'] == 'Garfield-Dan Ryan']
Garfield_South_Elevated = df[df['stationname'] == 'Garfield-South Elevated']

# Configurar el tamao de la figura
sns.set(rc={'figure.figsize':(11, 4)})

# Graficar los datos de Garfield-Dan Ryan
plt.figure(figsize=(11, 4))
plt.plot(Garfield_Dan_Ryan.index, Garfield_Dan_Ryan['rides'], linewidth=0.5, label='Garfield-Dan Ryan')
plt.title('Rides en Garfield-Dan Ryan')
plt.xlabel('Fecha')
plt.ylabel('Nmero de Rides')
plt.grid(True)
plt.legend()
plt.show()

# Graficar los datos de Garfield-South Elevated
plt.figure(figsize=(11, 4))
plt.plot(Garfield_South_Elevated.index, Garfield_South_Elevated['rides'], linewidth=0.5, label='Garfield-South Elevated')
plt.title('Rides en Garfield-South Elevated')
plt.xlabel('Fecha')
plt.ylabel('Nmero de Rides')
plt.grid(True)
plt.legend()
plt.show()
```
